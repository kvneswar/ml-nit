```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('data.csv')

print(df)

sns.displot(df.InternetUsers)

# plot the graph using one variable - UNIVARITE analysis
# plot the graph using two variable - BIVARITE analysis
# plot the graph using morethan two variable - MULTIVARITE analysis
# OUTLIER: statistic outlier is the data which is significantly very far from other observation
```

```python
sns.displot(df.InternetUsers, bins=20)
```

```python
sns.boxplot(data=df, x="IncomeGroup", y='BirthRate')
```

```python
sns.lmplot(data=df, x="InternetUsers", y='BirthRate')
```

```python
sns.lmplot(data=df, x="InternetUsers", y='BirthRate', fit_reg=False)
```

```python
sns.lmplot(data=df, x="InternetUsers", y='BirthRate', fit_reg=True)
```

```python
sns.lmplot(data=df, x="InternetUsers", y='BirthRate', fit_reg=False, hue='IncomeGroup')
```
